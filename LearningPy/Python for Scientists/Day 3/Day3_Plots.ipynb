```python
import numpy as np

df = pd.read_csv('../Desktop/1_rms.log','\s+', names=['frames','rms']).drop([0,1])
df = df.astype(float)
```

```python
plt.plot(df.rms)
plt.legend()
```

```python
theta = np.array(range(len(df.rms)))*(2*np.pi)/(len(df.rms))
rms= list(df.rms)
```

```python
plt.polar(theta, list(df.rms))
```

```python
r = df.rms.rolling(200)
r
```

```python
r.mean()
```

```python
plt.plot(df.rms)
plt.plot(r.mean(),c = 'r')
plt.legend()
```

```python
plt.polar(theta, rms)
plt.polar(theta, r.mean(), color='r', label='rms_mean')
plt.legend()
```

```python
plt.figure(figsize=(10,8))
plt.polar(theta, rms)
plt.polar(theta, r.mean(), color='r', label='rms_mean')
plt.legend(bbox_to_anchor=(1, 1))
```

```python
plt.figure(figsize=(10,8), dpi=200, facecolor='gray')
plt.polar(theta, r.mean(), color='b', label='rms_mean')
plt.legend(bbox_to_anchor=(1.5, 1))
plt.polar(theta, rms, c='b', alpha=0.5)
```
