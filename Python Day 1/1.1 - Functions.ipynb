### Functions. Exemple

Look at the argument show. This argument is set in False by default, meaning that the if steatment in the function will fail and execute the else steatment. But if you pass the 'show=True' as an argument, the show value will change and the code executed by the function. 

```python
def word_information(word, data=False):
    if data:
        vocals = [x for x in word if x in ['a','e','i','o','u']]
        consonants = [x for x in word if x not in ['a','e','i','o','u']]
        print('The word has {} vocals and {} consonants'.format(len(vocals), len(consonants)))
        print('\nVocals: ' + ','.join(vocals) + '\nConsonants: ' + ','.join(consonants))
        return vocals, consonants
    else:
        vocals = [x for x in word if x in ['a','e','i','o','u']]
        consonants = len(word) - len(vocals)
        print('The word has {} vocals and {} consonants'.format(len(vocals),consonants))
```
